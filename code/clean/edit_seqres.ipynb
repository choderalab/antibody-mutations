# Functions

```python
def delete_lines(lines, seqres_records, seqres_chains, seqres_numres, offset_start, offset_end):
"""Deletes residues as specified by SEQRES line indices, chain, and residue indices.

Parameters
----------
lines : list of strings
    The lines read in from the PDB file
seqres_records : list of ints
    The SEQRES line index(es) to remove
seqres_chains : list of strings
    The chain id(s) from which to remove SEQRES sequences
seqres_numres : int
    The number of residues in SEQRES chain after removing desired sequence
offset_start : int
    The index of first residue in first seqres_record to delete, starts from index 0
offset_end : int
    The index of one after the last residue (in the last seqres_record) to include in segment to delete
    
Returns
-------
lines_deleted : list of strings
    The lines with deleted residues from SEQRES
"""
    lines_deleted = []
    for line in lines:
        record_name = line[0]
        if record_name == "SEQRES":
            temp_line = [element for element in line if element != '']
            record_number = int(temp_line[1])
            chain = temp_line[2]
            if chain in seqres_chains:
                # Edit numRes in SEQRES record
                if line[1] == '' and line[2] == '':
                    line[6] = str(seqres_numres)
                else:
                    line[5] = str(seqres_numres)

                # Delete residues in specific seqres_records for seqres_chains
                if record_number in seqres_records:
                    if record_number == min(seqres_records):
                        line = line[:8+offset_start] if line[1] == '' and line[2] == '' else line[:7+offset_start]
                        if len(line) < 9:
                            continue # Do not append line
                    elif record_number == max(seqres_records):
                        line = line[:8] + line[8+offset_end:] if line[1] == '' and line[2] == '' else line[:7] + line[7+offset_end:]
                        if len(line) < 9:
                            continue # Do not append line
                    else:
                        # Do not append line
                        continue
        lines_deleted.append(line)
    return lines_deleted
```

```python
def split_chains(lines, seqres_line, seqres_chains, seqres_numres_old, seqres_numres_new, offset, d_chains):
"""Splits SEQRES chain as specified by SEQRES line indices, chain, and residue indices.
Note: This maintains chain ids (H-L-F becomes H-L-F-X after split).

Parameters
----------
lines : list of strings
    The lines read in from the PDB file
seqres_line : int
    The line from which to split the SEQRES sequences
seqres_chains : list of strings
    The chain id(s) from which to split SEQRES sequences
seqres_numres_old : int
    The number of residues in first half of split SEQRES chain 
seqres_numres_new : int
    The number of residues in latter half of SEQRES chain
offset : int
    The index of first residue after the split, starts from index 0
d_chains : dict
    Contains mapping of chain ids, where key : old chain id to split + "_new", value : new chain id

Returns
-------
lines_split : list of strings
    The lines with chains split in SEQRES
"""
    lines_split = []
    new_chain_added = False
    previous_chain = ""
    current_chain = ""
    for line in lines:
        record_name = line[0]
        if record_name == "SEQRES":
            record_number = int([element for element in line if element != ''][1])
            
            # Save chain id
            if line[1] == '' and line[2] == '':
                chain = line[4]
            else:
                chain = line[3]
            
            current_chain = chain
            if current_chain != previous_chain:
                new_chain_added = False
            
            if chain in seqres_chains:
                # If line is the designated line to be split, split it into two lines with the second line in new chain
                if record_number == seqres_line:
                    line_copy = line
                    line = line[:8+offset] if line[1] == '' and line[2] == '' else line[:7+offset]
                    next_line = line_copy[:8] + line_copy[8+offset:] if line_copy[1] == '' and line_copy[2] == '' else line_copy[:7] + line_copy[7+offset:]
                    
                    # Change numres of old chain
                    if line[1] == '' and line[2] == '':
                        line[6] = str(seqres_numres_old) 
                        if seqres_numres_old < 100: # Add a space before the numres if it only has 2 digits
                            line = line[:6] + [''] + line[6:]
                    else:
                        line[5] = str(seqres_numres_old)
                        if seqres_numres_old < 100: # Add a space before the numres if it only has 2 digits
                            line = line[:5] + [''] + line[5:]
                                
                    # Change chain id and numres of new chain
                    if next_line[1] == '' and next_line[2] == '':
                        next_line[4] = d_chains[chain]
                        next_line[6] = str(seqres_numres_new)
                        if seqres_numres_new < 100: # Add a space before the numres if it only has 2 digits
                            line = line[:6] + [''] + line[6:]
                    else:
                        next_line[3] = d_chains[chain]
                        next_line[5] = str(seqres_numres_new)
                        if seqres_numres_new < 100: # Add a space before the numres if it only has 2 digits
                            line = line[:5] + [''] + line[5:]
                    
                    # Remove '' after residues in the after split line
                    while next_line[len(next_line)-1] == '':
                        next_line = next_line[:len(next_line)-1]
                    
                    lines_split.append(line)
                    lines_split.append(next_line)
                    new_chain_added = True
                    continue
                
                
                # For the lines in the second half of the split chain, change chain id and numres
                if new_chain_added:
                    if line[1] == '' and line[2] == '':
                        line[4] = d_chains[chain]
                        line[6] = str(seqres_numres_new)
                        if seqres_numres_new < 100: # Add a space before the numres if it only has 2 digits
                            line = line[:6] + [''] + line[6:]
                    else:
                        line[3] = d_chains[chain]
                        line[5] = str(seqres_numres_new)
                        if seqres_numres_new < 100: # Add a space before the numres if it only has 2 digits
                            line = line[:5] + [''] + line[5:]
                            
                else: # For the lines in the first half of the split chain, change numres
                    if line[1] == '' and line[2] == '':
                        line[6] = str(seqres_numres_old) 
                        if seqres_numres_old < 100: # Add a space before the numres if it only has 2 digits
                                line = line[:6] + [''] + line[6:]
                    else:
                        line[5] = str(seqres_numres_old)
                        if seqres_numres_old < 100: # Add a space before the numres if it only has 2 digits
                                line = line[:5] + [''] + line[5:]
            
            previous_chain = current_chain
        lines_split.append(line)
    
    return lines_split
```

```python
def fill_lines(lines_deleted):
'''Fills in incomplete SEQRES lines after deletion of residues. E.g. if one line no only has 5 residues.

Parameters
----------
lines_deleted : list of strings
    The lines outputted by delete_lines
    
Returns
-------
lines : list of strings
    The lines with SEQRES lines filled where necessary.
'''

    lines = []
    previous_line = lines_deleted[0]
    for line in lines_deleted:
        record_name = line[0]
        previous_record_name = previous_line[0]
        if record_name == "SEQRES":
            # Get number of residues, chain, and number of missing residues of previous line
            previous_line_residues = [res for res in previous_line[8:] if res != ''] if previous_line[1] == '' and previous_line[2] == '' else [res for res in previous_line[7:] if res != '']
            previous_line_chain = previous_line[4] if previous_line[1] == '' and previous_line[2] == '' else previous_line[3]
            previous_line_missing_residues = 13 - len(previous_line_residues)
            
            # Get number of residues and chain of current line
            current_line_residues = [res for res in line[8:] if res != ''] if line[1] == '' and line[2] == '' else [res for res in line[7:] if res != '']
            current_line_chain = line[4] if line[1] == '' and line[2] == '' else line[3]
            
            # Check if there are missing residues and if previous chain matches the current chain
            # --> fill previous line with current residues
            if previous_record_name == "SEQRES" and previous_line_missing_residues > 0 and previous_line_chain == current_line_chain:
                # If there aren't enough residues to completely fill the previous line, don't update previous line or append current line
                if previous_line_missing_residues > len(current_line_residues):
                    previous_line = previous_line + current_line_residues
                    # Remove the previous line from lines and add the updated version of the previous line
                    lines = lines[:-1]
                    lines.append(previous_line)
                    continue
                else:
                    previous_line = previous_line + current_line_residues[:previous_line_missing_residues]
                    # Remove the previous line from lines and add the updated version of the previous line
                    lines = lines[:-1]
                    lines.append(previous_line)
                    # Update current line with residues added to previous line removed                    
                    line = line[:8] if line[1] == '' and line[2] == '' else line[:7]
                    if line[len(line)-1] != '':
                        line += ['']
                    line += current_line_residues[previous_line_missing_residues:]
        previous_line = line
        lines.append(line)
    return lines
            
```

```python
def reindex_lines(lines_filled):
'''Re-index SEQRES lines after filling in incomplete SEQRES lines. 
E.g. if one line has index 9 and the next line has index 12, 
this function sets the second line as index 10 and all subsequent line indexes in consecutive order.

Parameters
----------
lines_filled : list of strings
    The lines outputted by fill_lines()

Returns
-------
lines : list of strings
    The lines with SEQRES lines reindexed.

'''
    lines = []
    previous_line = []
    for line in lines_filled:
        record_name = line[0]
        if record_name == "SEQRES" and previous_line[0] == 'SEQRES':
            previous_line_chain = previous_line[4] if previous_line[1] == '' and previous_line[2] == '' else previous_line[3]
            current_line_chain = line[4] if line[1] == '' and line[2] == '' else line[3]
            if previous_line_chain == current_line_chain and previous_line[0] == 'SEQRES':
                previous_line_index = int(previous_line[3]) if previous_line[1] == '' and previous_line[2] == '' else int(previous_line[2])
                current_line_index = int(line[3]) if line[1] == '' and line[2] == '' else int(line[2])
                if current_line_index != previous_line_index + 1:
                    if line[1] == '' and line[2] == '':
                        if previous_line_index + 1 < 10:
                            line[3] = str(previous_line_index + 1)
                        else:
                            line[2] = str(previous_line_index + 1)
                            line = line[0:3] + line[4:]
                    else:
                        if previous_line_index + 1 < 10:
                            line[2] = ''
                            line = line[:3] + [str(previous_line_index + 1)] + line[3:]
                        else:
                            line[2] = str(previous_line_index + 1)
            else:  # Set first line at the start of each chain to index 1
                if line[1] == '' and line[2] == '':
                    line[3] = 1
                else:
                    line[2] = 1
                    line = line[0:3] + line[4:]
        previous_line = line
        lines.append(line)
    return lines
            
            
```

```python
def expand_seqres(lines):
'''Take 4jhw seqres (1 F protein and 1 antibody) and expand it to include 3 F proteins and 1 antibody

Parameters
----------
lines : list of strings
    The lines from the PDB.
    
Returns
-------
new_expanded : list of strings
    The lines from the PDB, with SEQRES expanded

'''
    new_lines = []
    F_lines = []
    X_lines = []
    for line in lines:
        record_name = line[0]
        if record_name == "SEQRES":
            # Save chain id
            if line[1] == '' and line[2] == '':
                chain = line[4]
            else:
                chain = line[3]
            if chain == 'F':
                F_lines.append(line)
            elif chain == 'X':
                X_lines.append(line)
            if len(X_lines) == 29: # After all SEQRES lines for chain X have been iterated through
                for line in F_lines: # Add chain A
                    if line[1] == '' and line[2] == '':
                        new_line = line[:4] + ['A'] + line[5:]
                    else:
                        new_line = line[:3] + ['A'] + line[4:]
                    new_lines.append(new_line) 
                for line in X_lines: # Add chain Y
                    if line[1] == '' and line[2] == '':
                        new_line = line[:4] + ['Y'] + line[5:]
                    else:
                        new_line = line[:3] + ['Y'] + line[4:]
                    new_lines.append(new_line) 
                for line in F_lines: # Add chain D
                    if line[1] == '' and line[2] == '':
                        new_line = line[:4] + ['D'] + line[5:]
                    else:
                        new_line = line[:3] + ['D'] + line[4:]
                    new_lines.append(new_line) 
                for line in X_lines: # Add chain Z
                    if line[1] == '' and line[2] == '':
                        new_line = line[:4] + ['Z'] + line[5:]
                    else:
                        new_line = line[:3] + ['Z'] + line[4:]
                    new_lines.append(new_line)
        new_lines.append(line)
    return new_lines
            
```
